# Pengubahan Tipe Data Kolom Transaksi

```python
import pandas as pd
df = pd.read_csv('/content/transaksi.csv')
df
```

```python
df['Harga Asli'] = df['Harga Asli'].replace('[Rp,]', '', regex=True).astype(float)
df['Tanggal Transaksi'] = pd.to_datetime(df['Tanggal Transaksi'], format='%Y-%m-%d')
df
```

```python
df1= df.groupby('Customer')['Harga Asli'].sum().reset_index()
df1['Harga Asli'] = df1['Harga Asli'].map('{:,.2f}'.format)
print(df1.head())
```
